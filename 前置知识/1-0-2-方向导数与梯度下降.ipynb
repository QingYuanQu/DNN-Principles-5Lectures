# 方向导数与梯度

## 方向导数

&emsp;&emsp;偏导数反映的是函数沿坐标轴方向的变化率，但许多物理现象告诉我们，只考虑函数沿坐标轴方向的变化率是不够的，例如,热空气要向冷的地方流动，气象学中就要确定大气温度、气压沿着某些方向的变化率，因此我们有必要来讨论函数沿任一指定方向的变化率问题。

设$l$是xy平面上以$P(x_0,y_0)$为始点的一条射线,$e_l=(cos\alpha,cos\beta)$是与$l$同方向的单位向量(图9-9).  
射线$l$的参数方程为

$$
\begin{equation}
\left\{
\begin{aligned}
x=x_0 + tcos ,\\
y=y_0 + tcos\beta\\
\end{aligned}
\right.
\end{equation}
(t>=0)
$$

&emsp;&emsp;设函数$z=f(x,y)$在点$P(x_0,y_0)$的某个邻域$U(P_0)$内有定义，$P(x+t \cos \alpha, y+ t \cos\beta)$为$l$上另一点，且$P \in U(P_0)$。如果函数增量$(x_0+t \cos \alpha, y_0 + t \cos \beta )-f(x_0, y_0)$与$P$到$P_0$的距离$|PP_0|=t$的比值

$$
\frac{f(x_0+tcos \alpha, y_0+tcos \beta) - f(x_0, y_0)}{t}
$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 创建画布
fig = plt.figure()
ax = fig.add_subplot(111)

# 射线 l
ax.quiver(1,2,5,5, angles='xy', scale_units='xy', scale=1 )
ax.text(6,7-0.4, r"$l$")

# 与射线同向的单位向量 el
ax.quiver(1,2,1,1, angles='xy', scale_units='xy', scale=1 )
ax.text(2-0.4,3,r"$e_l$")

# 横纵不缩放
ax.set_aspect(1)

# p0
ax.text(1-0.8,2-0.3,r"$P_0(x_0, y_0)$")

# P
plt.scatter([4],[5],color='k')
ax.text(4+0.2,5-0.1,r"$P(x, y)$")

# 单位向量 i, j 及 对应文本
ax.quiver(1,2,1,0, angles='xy', scale_units='xy', scale=1 )
ax.quiver(1,2,0,1, angles='xy', scale_units='xy', scale=1 )
ax.text(1.5,1.75,r"$i$", va="center", color="k")
ax.text(0.7,2.5, r"$j$", va="center", color="k")

# P在x,y方向上的投影
ax.quiver(1,2,3,0,  angles='xy', scale_units='xy', scale=1, color="r", alpha=0.2)
ax.quiver(1,2,0,3,  angles='xy', scale_units='xy', scale=1, color="r", alpha=0.2)

# alpha beta
ax.text(1+0.3,2+0.1,r"$\alpha$")
ax.text(1+0.05,2+0.3,r"$\beta$")

# 显示网格
ax.grid(axis="both", color='gray', linestyle='-', linewidth=0.5)  

# 设置坐标轴箭头样式
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# 不显示刻度
ax.set_xticks(range(0,8),[])
ax.set_yticks(range(0,8), [])

# x,y, label
ax.set_xlabel('x', rotation=0, loc="right")
ax.set_ylabel('y', rotation=0, loc="top")

# 设置坐标轴范围
plt.xlim(0, 7)
plt.ylim(0, 7)

# 显示图形
plt.show()
```

当$P$沿着$l$趋于$P_0$(即$t→0^+$)时的极限存在，那么称此极限为函数$f(x,y)$在点$P_0$沿方向$l$的**方向导数**，记作$\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)}$即

$$
\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)} = \lim_{t \to o^+} \frac{f(x_0+tcos \alpha, y_0+tcos \beta) - f(x_0, y_0)}{t} \tag{7-1}
$$

从方向导数的定义可知，方向导数$\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)}$就是函数$f(x,y)$在点$P_0(x_0,y_0)$处**沿$l$方向的变化率**.

若函数$f(x,y)$在点$P_0(x_0,y_0)$的**偏导数**存在，  
$e_l=i=(1,0)$，则

$$
\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)} = \lim_{t \to o^+} \frac{f(x_0+t, y_0) - f(x_0, y_0)}{t}  = f_x(x,y)
$$

又若$e_l=j=(0,1)$,则

$$
\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)} = \lim_{t \to o^+} \frac{f(x_0, y_0+t) - f(x_0, y_0)}{t}  = f_y(x_0, y_0)
$$

但反之,若$e_l=i$，$\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)}$存在，则$\left. \frac{\partial f}{\partial x} \right|_{(x_0,y_0)}$未必存在。

例如,$z = \sqrt{x^2+y^2}$，在点
$0(0,0)$处沿$l=i$方向的方向导数
$\left. \frac{\partial f}{\partial l} \right|_{(0,0)} = 1$
,而偏导数  $\left. \frac{\partial f}{\partial x} \right|_{(x_0,y_0)}$
不存在.

关于方向导数的存在及计算，我们有以下定理.  
**定理** 如果函数$f(x,y)$在点$P_0(x_0,y_0)$可微分，那么函数在该点沿任一方向$l$的方向导数存在，且有

$$
\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)} =  f_x(x_0,y_0) \cos \alpha + f_y(x_0,y_0) \cos \beta
$$

其中$\cos \alpha$和$\cos \beta$是方向$l$的方向余弦.

**证**

由假设 $f(x,y)$在点$(x_0, y_0)$可微分，故：

$$
\begin{align}
\Delta z = f(x_0 + \Delta x, y_0+\Delta y) - f(x_0, y_0) =f_x (x_0, y_0) \Delta x + f_y(x_0, y_0) \Delta y  + o(\sqrt{\Delta x ^2 + \Delta y^2} )
\end{align}
$$

但点$(x_0+\Delta x, y_0 + \Delta y)$在以$(x_0, y_0)$为始点的射线$l$上时， 应有$\Delta x = t \cos \alpha,  \Delta y = t \cos \beta , \sqrt{\Delta x ^2 + \Delta y^2} =t$

即：  
$$
\Delta z = f_x(x_0, y_0) t \cos \alpha + f_y(x_0, y_0) t \cos \beta + o(\sqrt{ ( t \cos \alpha)^2 + (t\cos \beta)^2} 
$$

由导数的定义知

$$
\begin{equation}
\lim_{t \to o^+} \frac{f(x_0+t\cos\alpha, y_0+t \cos\beta) - f(x_0, y_0)}{t}
\end{equation}
$$

由上述修改后的全微分公式知：
$$
\lim_{t \to 0+}  \frac{ f_x(x_0, y_0) t \cos \alpha + f_y(x_0, y_0) t \cos \beta + o(\sqrt{ ( t \cos \alpha)^2 + (t\cos \beta)^2}  }  {t}
$$

$$
\lim_{t \to 0+}  \frac{ f_x(x_0, y_0) t \cos \alpha + f_y(x_0, y_0) t \cos \beta + o(t) }  {t}
$$

$$
\lim_{t \to 0^+} \frac{f_x(x_0, y_0)t \cos \alpha}{t} + \lim_{t \to 0^+} \frac{f_y(x_0, y_0)t \cos \beta}{t} + \lim_{t \to 0^+} \frac{o(t)}{t}  
$$

$$
f_y(x_0, y_0) \cos \alpha+ f_x(x_0, y_0) \cos \beta
$$

## 梯度

与方向导数有关联的一个概念是函数的梯度，  

在二元函数的情形，设函数$f(x,y)$在平面区域$D$内具有一阶连续偏导数，则对于每一点$P_0(x_0,y_0) \in D $，都可定出一个**向量**

$$
f_x(x_0, y_0)i + f_y(x_0, y_0)j
$$

这**向量**称为函数$f(x,y)$在点$P_0(x_0,y_0)$的**梯度**，记作$gradf(x_0,y_0)$或$\nabla f(x_0,y_0)$，即
$$
grad f(x_0,y_0)=\nabla f(x_0,y_0) = f_x(x_0,y_0)i+f_y(x_0,y_0)j.
$$

### 梯度向量与方向导数

(1) 当$\theta=0$，即方向$e_l$，与梯度$\nabla(x_0,y_0)$的方向相同时，函数$f(x,y)$增加最快。此时，函数在这个方向的方向导数达到最大值，这个最大值就是梯度$\nabla f(x_0,y_0)$的模，即
$$
\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)}= |\nabla f(x_0,y_0)|.
$$

(2) 当$\theta=\pi$，即方向$e_l$，与梯度$\nabla(x_0,y_0)$的方向相反时，函数$f(x,y)$增加最快。此时，函数在这个方向的方向导数达到最小值，即
$$
\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)}= -|\nabla f(x_0,y_0)|.
$$

(3) 当$\theta=\frac{\pi}{2}$，即方向$e_l$与梯度$\nabla(x_0,y_0)$的方向正交时，函数的变化率为零，即
$$
\left. \frac{\partial f}{\partial l} \right|_{(x_0,y_0)}= |\nabla f(x_0,y_0)|\cos \theta = 0.
$$

上图，梯度向量在$x>0, y>0$空间的正方向区域内。  
如果当前点$(x_a, x_b)$在$x<0, y<0$区域内，  $i=(1,0), j=(0,1)$ 
- $f_x$随着$x$的增加，$z$值降低，$f_x<0$
- $f_y$随着$y$的增加，$z$值增加，$f_y>0$
- $\nabla = f_x i + f_y j$

## 梯度下降法


假设希望求解目标函数 $f(x)=f(x_{1},\cdots,x_{n})$ 的最小值，可以从一个初始点 ${x}^{(0)}=(x_{1}^{(0)},\cdots,x_{n}^{(0)})$ 开始，基于学习率$ \eta>0$ 构建一个迭代过程：当 $i\geq 0$ 时，

\begin{eqnarray*} x_{1}^{(i+1)} &=& x_{1}^{(i)} - \eta\cdot \frac{\partial f}{\partial x_{1}}({x}^{(i)}), \\ &\cdots& \\ x_{n}^{(i+1)} &=& x_{n}^{(i)} - \eta\cdot \frac{\partial f}{\partial x_{n}}({x}^{(i)}). \end{eqnarray*}

其中 ${x}^{(i)} = (x_{1}^{(i)},\cdots,x_{n}^{(i)})$ ，一旦达到收敛条件的话，迭代就结束。从梯度下降法的迭代公式来看，下一个点的选择与当前点的位置和它的梯度相关。反之，如果要计算函数 $f({x})=f(x_{1},\cdots,x_{n})$ 的最大值，沿着梯度的反方向前进即可，也就是说：

\begin{eqnarray*} x_{1}^{(i+1)} &=& x_{1}^{(i)} + \eta\cdot \frac{\partial f}{\partial x_{1}}({x}^{(i)}), \\ &\cdots& \\ x_{n}^{(i+1)} &=& x_{n}^{(i)} + \eta\cdot \frac{\partial f}{\partial x_{n}}({x}^{(i)}). \end{eqnarray*}

其中$ {x}^{(i)} = (x_{1}^{(i)},\cdots,x_{n}^{(i)})$ 。整体来看，无论是计算函数的最大值或者最小值，都需要构建一个迭代关系 $g $，那就是：

\begin{eqnarray*} {x}^{(0)} \stackrel{g}{\longrightarrow} {x}^{(1)} \stackrel{g}{\longrightarrow} {x}^{(2)} \stackrel{g}{\longrightarrow} \cdots, \end{eqnarray*}

也就是说对于所有的$ i\geq 0 $，都满足迭代关系$ x^{(i+1)}=g(x^{(i)})$ 。所以，在以上的两个方法中，我们可以写出函数 $g$ 的表达式为：

\begin{eqnarray*} g({x})=\begin{cases} {x} - \eta\nabla f({x}) &\text{梯度下降法},\\ {x} + \eta\nabla f({x}) &\text{梯度上升法}. \end{cases} \end{eqnarray*}

### 梯度下降可视化

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams["font.family"] = ["SimHei", "Times New Roman"]

# 创建3D图形
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# 定义二元函数
def f(x, y):
    return x**2 + 2*y**2  # 举例：这里定义了一个简单的二次函数

# 计算方向导数
def fxy(a, b):
    fx = 2 * a
    fy = 4 * b
    return fx, fy

# 生成数据
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 绘制二元函数曲面
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# (x0,y0)
x_0 = 1
y_0 = 1

# 点 (x0,y0,f(x0,y0))
ax.scatter(x_0, y_0, f(x_0,y_0), color="r")

# 偏导数 (x_0)
fx0, fy0 = fxy(x_0, y_0) 

#梯度向量
ax.quiver(1,1,f(1,1), fx0*0.3, 0, 0)
ax.quiver(1,1,f(1,1), 0, fy0*0.3, 0)
ax.quiver(1,1,f(1,1), fx0*0.3, fy0*0.3, 0, color="m")

# 学习率 0.1
eta = 0.1

####################### 第一次梯度下降 #############################

x_1 = x_0 -eta*fx0
y_1 = y_0 -eta*fy0
print(x_1, y_1)

# 新的位置
ax.scatter(x_1, y_1, f(x_1, y_1), color="r")

####################### 第二次梯度下降 #############################

fx_1, fy_1 = fxy(x_1, y_1)
x_2 = x_1 - eta * fx_1
y_2 = y_1 - eta * fy_1
ax.scatter(x_2, y_2, f(x_2, y_2), color="r")

####################################################

# 自定义坐标轴
# ax.quiver(0,0,0,0,0,12, color="k",arrow_length_ratio=0.1)
# ax.quiver(-2,0,0,4,0,0, color="k", arrow_length_ratio=0.1)
# ax.quiver(0,-2,0,0,4,0, color="k",arrow_length_ratio=0.1)
ax.quiver([0,-2.5,0],[0,0,-2.5],[0,0,0], [0,5,0], [0,0,5], [12,0,0], arrow_length_ratio=0.1, color="k") # 上述三行代码简写

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
ax.set_zlim(0,12)
ax.set_xlim(-2.5,2.5)
ax.set_ylim(-2.5,2.5)
# ax.set_aspect("equal")
# 设置标题
plt.title('梯度下降法')

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 创建3D图形
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(121, projection='3d')

# 定义二元函数
def f(x, y):
    return x**2 + 2*y**2  # 举例：这里定义了一个简单的二次函数

# 计算方向导数
def fxy(a, b):
    fx = 2 * a
    fy = 4 * b
    return fx, fy

# 生成数据
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 绘制二元函数曲面
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)
# (x0,y0)
x_current = 2
y_current = 2

# 偏导数 (x_0)

#梯度向量

# 学习率 0.1
eta = 0.1
theta = 1

# 记录下初始x y值
x_history = [x_current]
y_history = [y_current]

cnt = 0
while True:
    fx_current, fy_current = fxy(x_current, y_current) 
    ax.quiver(x_current, y_current, f(x_current, y_current), eta*fx_current, 0, 0, color="k")
    ax.quiver(x_current, y_current, f(x_current, y_current), 0, eta*fy_current, 0, color="k")

    ax.quiver(x_current, y_current, f(x_current, y_current), eta*fx_current, eta*fy_current, 0, color="g")
    ax.quiver(x_current, y_current, f(x_current, y_current), -eta*fx_current, -eta*fy_current, 0, color="m")
    ax.scatter(x_current, y_current, f(x_current, y_current), color="r")
    x_current = x_current -eta*fx_current
    y_current = y_current -eta*fy_current
    x_history.append(x_current)
    y_history.append(y_current)
    cnt+=1
    if np.sqrt(fx_current**2 + fy_current**2)<theta:
        break

# 自定义坐标轴
ax.quiver([0,-2.5,0],[0,0,-2.5],[0,0,0], [0,5,0], [0,0,5], [12,0,0], arrow_length_ratio=0.1, color="k") # 上述三行代码简写

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
ax.set_zlim(0,12)
ax.set_xlim(-2.5,2.5)
ax.set_ylim(-2.5,2.5)
# ax.set_aspect("equal")

cs = fig.add_subplot(122)
cs.contour(X, Y, Z)
cs.set_xlabel('x')
cs.set_ylabel('y')
cs.set_title('contour')

cs.scatter(x_history, y_history, f(np.array(x_history), np.array(y_history))*50, c='red', marker='.')
cs.set_aspect("equal")
plt.tight_layout()
plt.show()
```
