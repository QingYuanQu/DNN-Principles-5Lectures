# 一枚硬币引发的故事

## 1. 随机现象：“抛硬币的结果不确定”

**例子**：你拿出 1 枚均匀硬币，往上抛，落地后是 “正面” 还是 “反面”？

**对应概念**：结果无法提前确定（可能正、可能反），但抛 100 次会发现 “正面约 50 次、反面约 50 次”—— 这就是 “有统计规律的随机现象”。

## 2. 随机试验：“抛 1 次硬币这个动作”

**例子**：我们把 “抛 1 次硬币，观察落地结果” 定义为 1 次试验。

**对应概念**：这个试验满足 3 个条件：

* 可重复（能抛无数次）；

* 结果不确定（不知道是正还是反）；

* 所有可能结果已知（只有 “正面”“反面” 2 种）

  —— 这就是随机试验。

## 3. 样本空间：“抛硬币的所有可能结果”

**例子**：抛 1 次硬币，所有可能的结果只有 2 个：“正面” 和 “反面”。

**对应概念**：把这 2 个结果组成集合，记为 Ω = {正面，反面}—— 这个集合就是样本空间（简单说，就是 “所有可能结果的清单”）。

## 4. 随机事件：“我们关心的某个结果”

**例子**：抛 1 次硬币，你关心 “落地是正面”，或者 “落地是反面”。

**对应概念**：“正面” 是样本空间 Ω 的一个子集（{正面}），“反面” 也是一个子集（{反面}）—— 这些 “我们关心的部分结果” 就是随机事件，可简称为 “事件”。

## 5. 事件概率：“正面或反面发生的可能性”

**例子**：因为硬币均匀，抛 1 次落地后，“正面” 和 “反面” 发生的机会相等。

**对应概念**：我们用 “概率” 描述这种可能性 ——

*   “正面” 发生的概率是 0.5（或 50%），记为 P (正面)=0.5；
*   “反面” 发生的概率是 0.5（或 50%），记为 P (反面)=0.5。

## 6. 随机变量：“给硬币结果贴个数字标签”

**例子**：为了方便计算，你规定：“如果落地是正面，就记为 1；如果是反面，就记为 0”。

**对应概念**：这里的 “1” 和 “0” 就是随机变量取值，给随机变量定义个符号，比如记为 X

* 随机变量 X 把样本空间的 “文字结果”（正面 / 反面）映射成 “数字”（1/0）

  本质是 “给结果起个数字名字”，方便用数学分析。

## 7. 随机变量的值域：“随机变量的所有可能数字”

**例子**：延续上面的抛硬币随机变量 X（正面→1，反面→0），X 能取到的数字只有 2 个：0 和 1。

**对应概念**：这 2 个数字组成的集合 {0,1}，就是随机变量 X 的值域 —— 简单说，就是 “随机变量能取到的所有数字的清单”。

## 8. 概率分布：“随机变量的数字与概率的对应关系”

**例子**：还是抛硬币的 X（0 = 反面，1 = 正面），我们列出 “X 的数字” 和 “对应概率”：

| 随机变量 X   | 0（反面） | 1（正面） |
| -------- | ----- | ----- |
| 概率 $P(X)$ | 0.5   | 0.5   |

**对应概念**：这个 “数字 + 概率” 的对应表，就是 X 的概率分布 —— 它完整描述了 “随机变量每个数字发生的可能性”。



## 9. 信息熵：“抛硬币结果的‘不确定性’有多大”

#### **例子**：

延续抛硬币的随机变量 X（0=反面，P=0.5；1=正面，P=0.5）—— 当你没看到结果时，你会好奇“这个结果能告诉我多少信息？”，以及“这个‘不确定的结果’，整体不确定程度有多大？”  

#### 1. 先看单个结果的信息量：

   1. 如果落地是“正面”（X=1），我们只需要用1个二进制位（比如“1”）就能记录这个结果，这意味着这个结果携带的信息量是1“比特（bit）”；
   2. 同理，“反面”（X=0）也只需要1个二进制位记录，信息量也是1bit。  

#### 2. 为什么是1bit？

   1. 因为硬币均匀时，“正面”或“反面”的概率都是 1/2，
   2. 我们用公式 **-log₂(1/概率)** 计算单个结果的信息量：
   3. 代入概率 1/2，得到 -log₂(1/(1/2)) = -log₂2 = 1bit，刚好和实际记录所需的二进制位数一致。  

#### 3. 再看“平均不确定性”：

   1. 我们需要的不是单个结果的信息量，而是“每次抛硬币前，结果的整体不确定程度”—— 这就需要把两个结果的信息量，按它们的概率“平均一下”。
   2. 计算过程很简单：（“反面”的信息量×其概率）+（“正面”的信息量×其概率），即（1bit×0.5）+（1bit×0.5）= 1bit。这个“平均后的信息量”，就是抛硬币结果的**信息熵**。



**对应概念**：  

1. 先明确“单个事件的信息量”：
   1. 对于随机变量 $X$ 的某个取值 $x$（比如 $ X=1 $），若其发生概率为 $P(x)$，则这个事件的信息量为 
   
      $I(x) = -log₂\frac{1}{P(x)}$（也等价于 $-\log₂P(x)$），单位是“比特（$bit$）”。  
   
   2. 这个公式的直觉是：概率越小的事件，发生时带给我们的“意外感”越强，信息量也越大（比如“抛硬币连续10次正面”概率极小，信息量远大于1bit）；而必然发生的事件（P=1），信息量为 0（因为毫无意外）。  
   
2. 再定义“信息熵”：
   1. 信息熵是随机变量 $X$ **所有可能取值的信息量，按各自概率加权平均**的结果，它衡量的是“随机变量整体的不确定性大小”，公式记为：  
$$
H(X)=-\Sigma[P(x) × \log_2P(x)]
$$
   2. 对抛硬币的例子代入计算：X 只有 0 和 1 两个取值，P(0)=0.5，P(1)=0.5，所以 H(X) = -[0.5×log₂0.5 + 0.5×log₂0.5] = -[0.5×(-1) + 0.5×(-1)] = 1bit。  
   3. 这个结果的含义是：“抛均匀硬币”这个随机现象，其结果的不确定性用1bit就能量化——换句话说，我们只需要1个二进制位的信息，就能消除它的不确定性。

## 12. 联合事件：“同时看两次抛硬币的结果”

当我们从“抛1次硬币”升级到“抛2次硬币”，就需要关注“两次结果同时发生”的情况，这就对应联合事件和联合概率。

**例子**：抛2次均匀硬币，定义两个随机变量：

- $ X $：第一次抛硬币的结果（0=反面，1=正面）
- $ Y $：第二次抛硬币的结果（0=反面，1=正面）  
我们关心“第一次正面且第二次正面”（即 $ X=1 $ 且 $ Y=1 $）、“第一次反面且第二次正面”（$X=0$ 且 $ Y=1 $）这类“同时发生的结果”——这些就是**联合事件**。

## 13. 联合概率

**对应概念**：

- 联合事件的概率称为**联合概率**，记为 $P(X=x, Y=y)$（读作“X取x且Y取y的概率”）。
- 因为硬币均匀且两次抛投独立（第一次结果不影响第二次），所以所有联合概率都相等：  
  比如 $P(X=1, Y=1) = P(X=1) \times P(Y=1) = 0.5 \times 0.5 = 0.25$，同理 $P(X=0,Y=0)、P(X=0,Y=1)、P(X=1,Y=0)$ 也都是0.25。
- 用表格更清晰展示“两次抛硬币”的联合概率分布：

| 联合事件（X,Y） | (0,0)（两次反） | (0,1)（第一次反、第二次正） | (1,0)（第一次正、第二次反） | (1,1)（两次正） |
|----------------|----------------|----------------------------|----------------------------|----------------|
| 联合概率 $P(X,Y)$ | 0.25           | 0.25                       | 0.25                       | 0.25           |

---

> 我们可以看出传达出抛两枚硬币的结果只需要2bit。

因此，对于三枚硬币的结果更多，8种结果，不确定性更大，传递出三枚硬币结果需要3bit；

所以为了衡量这种不确定性，我们可以取对数。

**结论：** 对于一个具有$N$个等概率结果的系统，其内在的不确定性可由$log₂(N)$量化。

**对数函数在此的出现并非偶然，它源于对二进制编码长度最优化问题的抽象。**



## 14. 联合熵：“两次抛硬币的‘总不确定性’有多大”

联合熵是“多个随机变量同时发生时的总不确定性”，对应之前的“联合概率”。

**例子**：抛2次均匀硬币（$ X $是第一次，$ Y $是第二次，联合概率都是0.25）。 
你在抛之前，对“两次结果的组合”（比如(正,反)、(反,正)）的不确定性，就是联合熵要描述的内容。

**对应概念**：

- 两个离散随机变量 $ X $ 和 $ Y $ 的联合熵公式为： 
  $ H(X,Y) = -\sum_{x}\sum_{y} P(X=x,Y=y) \log_2 P(X=x,Y=y) $ 
  （双重求和：遍历所有“X的取值×Y的取值”的组合）。
- 计算“两次均匀硬币”的联合熵： 
  总共有4种联合事件，每种概率0.25，所以： 
  $ H(X,Y) = -4 \times [0.25 \times \log_2 0.25] = -4 \times [0.25 \times (-2)] = 2 $ 比特。 
  这也符合直觉：两次独立的均匀硬币，总不确定性是“1次的熵（1比特）+ 1次的熵（1比特）”= 2比特。

联合熵描述了两个变量的总不确定性，而如果我们只想关注其中一个变量的不确定性，就需要用到‘边缘熵’——它基于边缘概率计算，是多变量中单个变量的不确定性。”



## 15. 边缘概率：“只关心第一次抛硬币的结果，忽略第二次”
当我们有多个变量的联合概率时，若“只关注其中一个变量，忽略其他变量”，就需要通过“求和”得到这个变量的概率——这就是边缘概率。它是“从多变量回归单变量”的桥梁。

**例子**：还是抛2次均匀硬币（$ X $=第一次，$ Y $=第二次），我们现在只关心“第一次抛到正面（$ X=1 $）”的概率，不管第二次（$ Y $）是正还是反。

**对应概念**：

- 边缘概率是对“不关心的变量所有可能取值”求和，公式为： 
  对 $ X $ 的边缘概率：$ P(X=x) = \sum_{y} P(X=x, Y=y) $（遍历 $ Y $ 的所有取值，求和联合概率）； 
  对 $ Y $ 的边缘概率：$ P(Y=y) = \sum_{x} P(X=x, Y=y) $（遍历 $ X $ 的所有取值，求和联合概率）。
- 用“补充边缘的联合概率表”直观展示计算过程（在原有联合表基础上，增加“行求和”和“列求和”，即边缘概率）：

|$Y\setminus X$|$X=0$|$ X=1 $| **边缘概率 $P(Y=y)$**|
|-------------|----------------|-------------|---------------------|
| $Y=0$| 0.25 (0,0) | 0.25 (1,0)  | $P(Y=0)=0.5$ |
| $ Y=1 $ | 0.25 (0,1) | 0.25 (1,1)  | $P(Y=1)=0.5$ |
| **边缘概率 $P(X=x)$** |$P(X=0)=0.5$|$P(X=1)=0.5$ | ——（总概率和为1）|

*   **核心直觉**：边缘概率就是“忽略不关心的变量，把所有包含目标变量的联合事件概率加起来”。比如求“第一次正（$ X=1 $）”的概率，不管第二次是正还是反，只要第一次是正，就把这两个联合事件（(1,0)、(1,1)）的概率相加，结果还是0.5，和单抛1次硬币的概率一致——这说明“多变量系统中，单个变量的边缘概率和它单独存在时的概率相同（独立情况下）”。






## 16. 边缘熵：“只看第一次抛硬币的不确定性，忽略第二次”
边缘熵是“边缘概率对应的不确定性”——即多变量系统中，**单个变量本身的不确定性**（与该变量单独存在时的熵完全一致），是从“总不确定性（联合熵）”回归“单变量不确定性”的概念。

**例子**：还是抛2次均匀硬币，我们现在想知道“只看第一次抛硬币（$ X $）的不确定性有多大”，不管第二次（$ Y $）的结果——这就是 $ X $ 的边缘熵。

**对应概念**：

- 边缘熵的计算完全依赖边缘概率，公式与单变量熵一致： 
  对 $ X $ 的边缘熵：$ H(X) = -\sum_{x} P(X=x) \log_2 P(X=x) $； 
  对 $ Y $ 的边缘熵：$ H(Y) = -\sum_{y} P(Y=y) \log_2 P(Y=y) $。
- 用“抛2次硬币”计算边缘熵： 
  从之前的边缘概率可知，$ X $ 的边缘概率是 $ P(X=0)=0.5 $、$ P(X=1)=0.5 $，所以： 
  $ H(X) = -[0.5\log_2 0.5 + 0.5\log_2 0.5] = 1 $ 比特。 
  同理，$ Y $ 的边缘熵 $ H(Y) = 1 $ 比特——和单抛1次均匀硬币的熵完全相同。
- **与联合熵的对比**：联合熵 $ H(X,Y)=2 $ 比特（总不确定性），边缘熵 $ H(X)=1 $ 比特（仅 $ X $ 的不确定性），这说明“联合熵是所有变量边缘熵的和（独立变量情况下）”——因为两次硬币独立，总不确定性=第一次的不确定性+第二次的不确定性。




## 17. 条件概率：“已知第一次结果，求第二次的概率”
条件概率是“在某个事件已经发生的前提下，另一个事件发生的概率”——正好呼应你最初问的“第一次为阳，三次皆阳”的逻辑，用抛硬币更易理解。

**例子**：还是抛2次硬币（$ X $ 是第一次，$ Y $ 是第二次）。现在问：“已知第一次抛到了正面（$ X=1 $ 已经发生），第二次也抛到正面（$ Y=1 $）的概率是多少？”

**对应概念**：

- 这种“已知X=1，求Y=1”的概率就是**条件概率**，记为 $P(Y=1 \mid X=1)$（读作“在X=1的条件下，Y=1的概率”）。
- 计算逻辑：“已知X=1”意味着“样本空间缩小了”——原来的4种联合事件，现在只保留“X=1”的2种：(1,0)和(1,1)，这2种情况的概率总和是 $0.25 + 0.25 = 0.5$。  
  其中“Y=1”的情况是(1,1)，概率是0.25，所以：  
  $P(Y=1 \mid X=1) = \frac{联合概率P(X=1,Y=1)}{已知事件的概率P(X=1)} = \frac{0.25}{0.5} = 0.5$。
- 核心公式：对任意两个事件，条件概率满足 $P(A \mid B) = \frac{P(A,B)}{P(B)}$（前提是 $P(B) \neq 0$）。




## 18. 条件熵：“已知第一次结果后，第二次的‘剩余不确定性’有多大”
条件熵是“在已知一个随机变量结果的前提下，另一个随机变量的剩余不确定性”——对应之前的“条件概率”，也是你最初关注的“条件熵”的核心。

**例子**：还是抛2次均匀硬币（$ X $第一次，$ Y $第二次）。现在问：“如果我已经知道第一次抛到了正面（$ X=1 $），那么第二次结果（$ Y $）的不确定性还有多大？”

**对应概念**：

- “已知X时Y的条件熵”记为 $H(Y \mid X)$，公式是“先对每个X的取值算‘条件下Y的熵’，再用X的概率加权平均”：  
  $H(Y \mid X) = \sum_{x} P(X=x) \times [-\sum_{y} P(Y=y \mid X=x) \log_2 P(Y=y \mid X=x)]$。
- 计算“两次均匀硬币”的 $H(Y \mid X)$：  
  因为两次独立，不管X取0还是1，$P(Y=0 \mid X=x)=0.5$，$P(Y=1 \mid X=x)=0.5$，所以“每个X条件下Y的熵”都是1比特。  
  又因为 $P(X=0)=P(X=1)=0.5$，所以：  
  $H(Y \mid X) = 0.5 \times 1 + 0.5 \times 1 = 1$ 比特。  
  这意味着：知道第一次结果后，第二次的不确定性还是1比特（因为两次独立，第一次结果没提供第二次的信息）。

## 21. 互信息：“第一次结果能‘减少’第二次的不确定性多少”
互信息衡量“一个随机变量的信息，能对另一个随机变量的不确定性减少多少”——本质是“熵与条件熵的差值”。

**例子**：抛2次硬币，若两次独立（比如正常抛），第一次结果（X）完全不能减少第二次（Y）的不确定性；但如果是“粘住的硬币”（第一次正→第二次必正，第一次反→第二次必反），X就能完全消除Y的不确定性。

**对应概念**：

- 互信息 $I(X;Y)$ 的核心公式（也是直觉含义）：  
  $I(X;Y) = H(Y) - H(Y \mid X)$  
  （“Y原本的不确定性”减去“知道X后Y的剩余不确定性”，就是X能给Y减少的不确定性）。
- 计算“两次独立均匀硬币”的互信息：  
  已知 $H(Y)=1$ 比特（Y本身是均匀硬币），$H(Y \mid X)=1$ 比特（知道X后Y的不确定性不变），所以：  
  $I(X;Y) = 1 - 1 = 0$ 比特。  
  这符合直觉：独立的两个变量，互信息为0（彼此不能提供信息）。
- 计算“粘住的硬币”（X=1→Y=1，X=0→Y=0）的互信息：  
  先算 $H(Y)$：Y和X完全一样，是均匀硬币，所以 $H(Y)=1$ 比特；  
  再算 $H(Y \mid X)$：知道X后，Y的结果完全确定（比如X=1→Y=1），所以“条件熵为0比特”；  
  因此 $I(X;Y) = 1 - 0 = 1$ 比特（X能完全消除Y的不确定性，互信息等于Y的初始熵）。
